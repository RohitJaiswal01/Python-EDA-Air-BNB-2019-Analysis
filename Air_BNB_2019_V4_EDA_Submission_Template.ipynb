## **Outlier Treatement**

````python
# IQR --> Q3-Q1 --> 75%-25%
# Q1 of minimum_nights
# Q3 of minimum_nights
Q1 = 1
Q3 = 5

IQR = Q3-Q1  # Answer is IQR = 4
print(f"Iqr is {IQR}")

# lower fence = Q1-1.5*IQR  --> 3-1.5(4) --> 3-6 --> -3  --> Negative Qutliers
# Hiher fence = Q3+1.5*IQR --> 7+1.5(4) --> 7+6 --> 13 --> Positive Outliers

lower_fence =  Q1 - 1.5*IQR
Higher_fence = Q3 + 1.5*IQR

print(f"Lower Fence is {lower_fence}")
print(f"Higher Fence is {Higher_fence}")
````
